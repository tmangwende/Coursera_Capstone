Import all the relevant libraries.

```
#dowmload all the required modules(dependencies)
import numpy as np #data manipulation in vectorised manner
import pandas as pd #data analysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import json #handle json files
!pip install geopy
from geopy.geocoders import Nominatim #converts address to latitute and longitude values
import requests #handle requests sending to API
from pandas.io.json import json_normalize #transforms JSON file to pandas dataframe
#matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors
#import k-means from clustering stage
from sklearn.cluster import KMeans
!pip install folium 
import folium #map rendering
print('Libraries imported.')
```

```
#use geopy to get latitude and longitude values of Toronto City in Ontario
#define an instance of the geocoder by defining a user_agent named on_explorer
address = 'Toronto City, ON'
geolocator = Nominatim(user_agent="on_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geographical coordinates of Toronto City are {},{}.'.format(latitude, longitude))
```

Create the map of Toronto City

```
#create map of Toronto given latitude and longitude values
map_torontocity = folium.Map(location=[latitude, longitude], zoom_start=10)
#add markers to the map
for lat, lng, borough, neighbourhood in zip(df_6['Latitude'], df_6['Longitude'], df_6['Borough'], df_6['Neighbourhood']):
  label = '{}, {}'.format(neighbourhood, borough)
  label = folium.CircleMarker([lat, lng], radius=5, popup=label, color='blue', fill=True, fill_color='#3186cc', fill_opacity=0.7, parse_html=False).add_to(map_torontocity)
#display map of Toronto City
map_torontocity
```
